```python
import pandas as pd

whr2021 = pd.read_csv('./world-happiness-report-2021.csv')
whr = pd.read_csv('./world-happiness-report.csv')
```

```python
# 1. ¿Cuál es el país más “feliz” del 2021 según la data?

MaxHap21 = whr2021['Ladder score'].max()
FMaxHap = whr2021[whr2021['Ladder score'] == MaxHap21]

MaxHapS21 = FMaxHap['Country name'].values[0]

print(f'El país más "feliz" del 2021 es: {MaxHapS21}')
```

```python
# 2.¿Cuál es el país más “feliz” del 2021 por continente según la data?

max_reg = whr2021.groupby('Regional indicator')['Ladder score'].max().reset_index()
resultado = whr2021.merge(max_reg, on=['Regional indicator', 'Ladder score'], how='inner')

print(resultado[['Regional indicator', 'Country name', 'Ladder score']])
```

```python
# 3 ¿Cuál es el país que más veces ocupó el primer lugar en todos los años?

max_year = whr.groupby('year')['Life Ladder'].idxmax()

reg_max_h = whr.loc[max_year]

conteo_paises = reg_max_h['Country name'].value_counts().reset_index()

conteo_paises.columns = ['Country name', 'cantidad_max_bienestar']

conteo_paises = conteo_paises.sort_values(by='cantidad_max_bienestar', ascending=False)

print('Países con la mayor valoración de bienestar y su cantidad de veces:')
print(conteo_paises)
```

```python
# 4.¿Qué puesto de Felicidad tiene el país con mayor GDP del 2020?

whr2021['year'] = 2021

df_merge = pd.merge(whr2021, whr, on=['Country name', 'year'], how='outer')
df_ordenado = df_merge.sort_values(by='Life Ladder', ascending=False)
df_ordenado.to_csv('new_df_ord.csv', index=False)
df_ordenado = df_ordenado.reset_index(drop=True)

df_ordenado.head()
```

```python
df_2020 = df_ordenado[df_ordenado['year'] == 2020]
pais_max_GDP = df_2020.loc[df_2020['Log GDP per capita'].idxmax()]['Country name']
puesto = df_ordenado[df_ordenado['Country name'] == pais_max_GDP].index[0] + 1

print(f'En el año 2020, el país con el mayor GDP es {pais_max_GDP}, y ocupa el puesto #{puesto} en el ranking de felicidad.')
```

```python
# 5. ¿En que porcentaje a variado a nivel mundial el GDP promedio del 2020 respecto al 2021? ¿Aumentó o disminuyó?

promedio_2020 = df_ordenado[df_ordenado['year'] == 2020]['Log GDP per capita'].mean()
promedio_2021 = df_ordenado[df_ordenado['year'] == 2021]['Logged GDP per capita'].mean()

variacion = promedio_2021 - promedio_2020

if variacion > 0:
    cambio = "creció"
else:
    cambio = "descendió"

porcentaje_variacion = abs((variacion / promedio_2020) * 100)    

print(f'Promedio de crecimiento en 2020: {promedio_2020:.2f}')
print(f'Promedio de crecimiento en 2021: {promedio_2021:.2f}')
print(f'Variación del promedio de 2020 a 2021: {variacion:.2f} ({cambio} en un {porcentaje_variacion:.2f}%)')
```

```python
#6. ¿Cuál es el país con mayor expectativa de vide (“Healthy life expectancy at birth”)? Y ¿Cuánto tenia en ese indicador en el 2019?

print('2005 a 2019:')
life = whr['Healthy life expectancy at birth'].idxmax()
resultado = whr.loc[life, ['Country name', 'Healthy life expectancy at birth', 'year']].reset_index(drop=True)
print(resultado)

print('')
print('2021:')

life2021 = whr2021['Healthy life expectancy'].idxmax()
resultado = whr2021.loc[life2021, ['Country name', 'Healthy life expectancy']]
print(resultado.to_string())
```
